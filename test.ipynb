```python
stopword=[] #建立停用詞字典字串等等要匯入
with open("stopword.txt","r",encoding="UTF-8") as file:
    for d in file.readlines():
        d=d.strip() #移除字串頭尾的空白
        stopword.append(d)

stopword.append(" ")
```

```python
import jieba 
import unicodedata
jieba.load_userdict("dict.txt.big")#匯入big5字典，避免亂碼
namelist=list(data["name"])
temp = []#外
for sentence in namelist:
    sep=jieba.cut(sentence)
    ls=[]
    for i in sep:
        if i in stopword:
            continue
        a=unicodedata.normalize('NFKC', i)
        b=nosig(a)
        ls.append(b)
    temp.append(ls)

data["jieba"]=temp
```
